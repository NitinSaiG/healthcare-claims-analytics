# 01 — Exploration
Quick checks on exported CSV.

```python
import pandas as pd
df = pd.read_csv('data/clean/claims_for_bi.csv', parse_dates=['admission_date','discharge_date'])
len(df), df.head()
```

```python
import matplotlib.pyplot as plt
monthly = df.groupby(df['admission_date'].dt.to_period('M'))['payment_amount'].sum().reset_index()
monthly['admission_date'] = monthly['admission_date'].astype(str)
plt.figure()
plt.plot(monthly['admission_date'], monthly['payment_amount'])
plt.title('Monthly Payment Trend')
plt.xlabel('Month')
plt.ylabel('Total Payment')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
